**Problem Statement :**
If you do the rice cooking without ever stirring it, you will find out that your well-cooked and tasty rice shows a pattern of holes like the one in the picture. Can you suggest a physical explanation for the phenomenon? Identify the relevant parameters.

16h : First attempt at forming the rice holes, using $80\pm1$g of rice, $148\pm1$g of water, in a container with diameter $12.3\pm0.1$cm and height $6.2\pm0.1$cm, heated over medium heat on a hotplate during 13 minutes. (see [Rice table](../Rice_attempt_summary/rice_attemp_table.ipynb) for more detail of the parameter for each experiment)

First result:\
<img src="riz_basmati_1.jpeg" width="500"/>

**Observations :**
1. Few or no holes formed.
2. After stirring and scooping, the bottom remains wet.

**Hypotheses :**
1. Starch concentration may be too low in the water.
2. Without a lid, moisture is not retained, causing the top layer of rice to dry out
3. The hotplate temperature may be too high, accelerating evaporation before the starch can form binding
4. Grain shape may be too short or too thin, preventing the formation of a dense or compact structure.

10h-13h : Experimentations on the influence of bubbles in the formation of the holes

**1.$~$** Reference succesfull experiment\
Reference experiment was prepared using long-grain rice (Alnatura, Migros). The system was brought to boiling without a lid at maximum heat, then the heat was reduced to medium–low and a lid was added to sustain boiling : rice $61.7\pm 0.1$g, water $156.5\pm 0.1$g, cooking time $18$ min $46$ s (note: for this specific rice brand, soaking for few seconds before cooking already leads to the formation of rice agglomerates, which did not appear for the previous one).

<img src="long_grain_11.jpeg" width="400"/><img src="long_grain_12.jpeg" width="400"/>

**Observations**
1. Holes have been formed uniformly, which means that with these current parameters the experiment is succesfull. This allow us the vary one parameter and evaluate its influence on the result.


**2.$~$** Minimising bubble formation\
A test minimizing bubble formation was performed by significantly lowering the temperature after reaching boiling and removing the lid : rice $61.5\pm 0.1$g, water $158.7\pm 0.1$g, cooking time $20$ min $02$ s.

<img src="long_grain_21.jpeg" width="400"/><img src="long_grain_22.jpeg" width="400"/>

**Observations**
1. Significantly less holes, confirming the strong correlation between bubbles and holes formation

**3.$~$** Refraining the boiling\
A test without boiling was conducted at low temperature, with the water heated just below boiling and then maintained at reduced power : rice $61.6\pm 0.1$g, water $158.1 \pm 0.1$g, cooking time $40$ min $10$ s (long cooking timee because no boiling).

<img src="long_grain_31.jpeg" width="400"/><img src="long_grain_32.jpeg" width="400"/>

**Observations**
1. No significant difference with the previous experiment.

**Conclusions :** Boiling appears to be a key parameter in the formation of holes !

**Conclusions :** Maintaining a sustained boiling regime is clearly critical for hole formation, as even basmati rice—which initially did not exhibit the phenomenon clearly—was able to produce satisfactory results under appropriate boiling conditions. However, the results also indicate that the type of rice, as well as its mass, geometry, and concentration, influence the outcome. The experiments do not allow a clear conclusion to be drawn regarding the specific role of starch concentration on the observed effects.

**Observations :**

**Hypotheses :**
1. Holes consistently appear with this type of rice; however, rather than promoting their formation, risotto rice seems to inhibit the development of some holes. Because of its high starch content, starch accumulates in the water and forms a highly concentrated film at the surface. This likely increases the effective surface tension, preventing some bubbles from breaking through the surface and, consequently, inhibiting the formation of holes.

Sixth attempt (return on long grain but experimentation with boiling phase) : Parboiled long-grain rice, covered with lid but not continuously; heated at maximum power until boiling, then at medium–low heat. The total time is unclear because different phase of low, medium, without, with lid were experimented. The Goal was to understand the reaction of the system when influencing the boiling. Rice : $87.0\pm 0.1$g; water : $210.9\pm 0.1$g.

**Observations :**

As the water level gradually decreased, distinct bubbling regions appeared at the surface (areas with no bubbles and areas with active bubbling). At this stage, the lid was lifted to photograph these regions in order to later compare them with the final spatial distribution of holes. It was then observed that the surface bubbling progressively weakened and eventually disappeared. This is interpreted as a sudden decrease in surface temperature : boiling could no longer be sustained because bubbles rising from the bottom of the pan were unable to burst at a surface that had cooled significantly. After the lid was replaced, the system gradually reheated and new bubbles began to burst at the surface again. At that point, fewer potential holes—through which bubbles could escape—were observed to form.

**Hypotheses :**

The force that had previously pushed the rice grains apart around a “chimney” through which a bubble emerged was abruptly suppressed. Since the rice was still largely immersed in water, the grains were able to slide and collapse back, closing the hole once this force vanished. When the rice subsequently reached a temperature high enough to sustain boiling again, the potential holes had partially closed, and a significant fraction of the water had already evaporated. As a result, the upper rice layer began to dry and stick together, which hindered the formation of larger and more stable chimneys.

Implementation of the idea :\
When the water level dropped below the level of the rice, the pan was removed from the hotplate and subjected to small, abrupt motions in order to mechanically perturb the material. It was clearly observed that the upper layer remained essentially rigid and immobile, while the underlying layers were unstable and easily displaced or disrupted.

« Several studies use water as a musical instrument, creating a variety of **sounds depending on how the surface of the water is stroked**. Explain **how the sound is formed** and **investigate how the properties of the sound depend on the stroke techniques**. https://www.youtube.com/watch?v=i5ECtJvKd94 »

Videos sources about the water as a musical instrument, in the aim of **exploring the different existing techniques** (to **give ideas for the sound experimenting phase**) to create sound with water :

1. « Water music of Vanuatu » (link : https://www.youtube.com/watch?v=i5ECtJvKd94) :
Use of water/water surface as a percussion instrument by women in the north of the Vanuatu Islands. Percussion techniques that exploit the interaction between air, water, and hands: striking the surface, using the impact of droplets on the surface, creating waves, changing the shape of the impact surface, etc. The video highlights the connection between the interaction of water, the shape of the cavity, and the resulting sound.

2. « Tan Dun’s Water concerto » (link : https://www.youtube.com/watch?v=wMkv2DlpaqY) :
List of different ways to make water interact with other objects to produce different sounds (depending on the material of the object, its shape, etc.). Various techniques: tapping with fingers or the palm of the hand, flicking, letting droplets or a stream of water fall, resonating a sound on the water surface, striking a cup with its base or its opening, hitting the back of a wooden bowl/metal cones half-immersed in water, immersing a vibrating gong in water.

3. « Liquindi - Baka women water drumming » (link : https://www.youtube.com/watch?v=C7ba1CNOLiI) :
The Baka forest people using water as a percussion instrument, another line of research into techniques: differences in sound depending on the depth of immersion of the object (for example, the whole arm rather than just the hand).

Scientific studies sources in the aim of **understanding the physical phenomenon of sound generation through water-air-object interaction**, by investigating the **variables** and the useful **experimental devices/procedure** :

1. « Sound generation as drop falls on a water surface », V. E. Prokhorov and Yu. D. Chashechkin (link : https://link.springer.com/article/10.1134/S1063771011050137) :\
When a droplet falls onto a water surface, it generates sound impulses both in the air and underwater (the characteristics depend on the droplet’s size and velocity). Experiments combining synchronized sound recordings and video show that the sounds originate from gas cavities detaching from the underwater cavity, which deform and then evolve into stable elliptical or spherical shapes.

3. « Physically Based Models for Liquid Sounds », KEES VAN DEN DOEL (link : https://dl.acm.org/doi/abs/10.1145/1101530.1101554) :\
Proposal of a physical model of the sound of an isolated bubble to generalize it for creating complex liquid sounds.

4. « Low‐frequency underwater sound generation by impacting transient cylindrical water jets », Ali R. Kolaini, Ronald A. Roy, Lawrence A. Crum, Yi Mao (link : https://pubs.aip.org/asa/jasa/article/94/5/2809/968702/Low-frequency-underwater-sound-generation-by) : \
The impact of a water jet on a calm surface leads to the formation of bubble plumes (sources of acoustic emissions). High-speed imaging experiments show that detached plumes oscillate and produce low-frequency sounds (depending on plume size, jet velocity, and container geometry). The acoustic intensity is correlated with the potential energy of the water jet.

5. « Improved Water Sound Synthesis using Coupled Bubbles », Xue, Kangrui and Aronson, Ryan M. and Wang, Jui-Hsien and Langlois, Timothy R. and James, Doug L. (link : https://dl.acm.org/doi/abs/10.1145/3592424) :\
Sound synthesis of bubbles based on coupled oscillator models and optimized numerical methods. The results show a realistic reproduction of low-frequency-rich sounds for flows containing hundreds of thousands of bubbles.

6. « Splashes as Sources of Sound in Liquids », G. J. Franz (link : https://pubs.aip.org/asa/jasa/article/31/8/1080/718696/Splashes-as-Sources-of-Sound-in-Liquids) :\
Study of the mechanisms of sound production during the entry of objects into water. Splashes from droplets or various objects produce acoustic dipoles.

7. « Sound radiation by various types of laboratory breaking waves in fresh and salt water », Ali R. Kolaini (link : https://pubs.aip.org/asa/jasa/article/103/1/300/561329/Sound-radiation-by-various-types-of-laboratory) :\
Study of wave sounds associated with oscillations of individual bubbles or bubble clouds. Highlighting the impact of saltwater density on the sound, compared to freshwater.

8. « Gas Bubbles as Sources of Sound in Liquids », M. Strasberg (link : https://pubs.aip.org/asa/jasa/article/28/1/20/748779/Gas-Bubbles-as-Sources-of-Sound-in-Liquids)

9. « Amplitude of water pouring sound », Boudina, Mouad and Kim, Joonoh and Kim, Ho-Young (link : https://journals.aps.org/prfluids/abstract/10.1103/PhysRevFluids.8.L122002)

10. « Why do hot and cold water sound different when poured? », Xiaotian Bi
, Dike Su, Qianyun Zhou (link : https://arxiv.org/abs/2403.14740)


11. « Sound amplitude of discrete bubbles entrained by an impacting water stream », Filippo Nelli, Shuang Zhu, Andrew Ooi, Richard Manasseh (link : https://pubs.aip.org/asa/jasa/article/156/1/350/3303114)


12. « The Sound of Water: Inferring Physical Properties from Pouring Liquids », P. Bagad, M. Tapaswi, C. G. M. Snoek and A. Zisserman (link : https://ieeexplore.ieee.org/abstract/document/10889950)

In summary: there is quite a lot of documentation studying the generation of sound with water, but many of them focus on underwater sounds. I tried to gather those that could also investigate the case of sound in air, though it remains to be seen whether sound generated underwater would also fit the topic. There are numerous avenues to test different sound-producing techniques (some artists even specialize in this). The next step is to be able to record the sound properly, and perhaps even film the technique, to understand what produces a certain type of sound and through which interaction—most likely, multiple effects will overlap.

#### Day report 15/10/25

Series of small tests and conclusions of the day :

1. Ruler-Battery Test: A battery was placed at the end of a ruler, with the other end resting on a table, and an impulse was applied at the end where the battery was attached, in the hope of observing that vibrations would stop sooner for a fully charged battery. The battery was attached simply using tape. No notable difference was observed between a fully charged battery and battery number 15. However, the experiment was quite rudimentary (timing by eye with a stopwatch, tape itself may provide damping, and its placement changes the contact surface with air, thus affecting external damping). Given that the rebound experiment works well, it was thought that the best method to obtain measurable results would involve abrupt energy variations, hence the rigid coupling of the ruler and battery. To be tested: a more rigorous version of this experiment. With the available rudimentary materials, this should not be overly complicated.

2. Tuning Fork Test (alone): The setup was designed to mimic the one with the box as closely as possible, but without the box itself. The tuning fork was placed at the same distance from the same falling ball, and the microphone was positioned near the prongs to capture the true sound of the fork. Additionally, adhesive pads were placed between the handle of the tuning fork and its support. Since these pads are made of a soft material, they absorb transmitted vibrations and prevent the tuning fork from resonating with other parts of the setup, ensuring minimal coupling so that the recorded sound originates only from the prongs. The measured decay constant ($\tau$) was found to be approximately 40% larger than the $\tau$ values obtained with the box. This confirms that the tuning fork itself contributes the most. The difference could be explained either by less rigid coupling or by a relatively significant contribution from the box, but in any case, the battery cannot have a large enough effect.

3. Sound Impact analysis : The battery was struck while held fixed in order to analyze the resulting sound. The impact sound is extremely brief, making detailed analysis difficult. A potentially more feasible approach is to hold the battery fixed and measure the impact duration, which is a coarser measurement and therefore easier to obtain than analyzing frequencies or other fine details. One possible method is to use a rigid pendulum to strike the battery, controlling the strike velocity and determining the impact time from the sound. It can be hypothesized that a fully charged battery would have a longer impact duration.

5. It was observed that when the battery is rigidly fixed and struck at one end, vibrations are clearly transmitted to the opposite end. This is advantageous because, given the difficulty of characterizing these vibrations with a microphone, a piezoelectric sensor could be used instead. The challenge lies in producing reproducible impacts: striking the battery with an object results in different impacts for each battery, as each one interacts with the striking object differently depending on its elasticity (impact duration, etc.). This complicates precise characterization of the phenomenon. Nevertheless, useful information can still be obtained indirectly, as the vibrations transmitted to the opposite end are expected to vary between batteries. This approach is potentially feasible, but modeling the precise relationship between the impact conditions (e.g., pendulum speed before collision) and the resulting vibrations will be challenging.

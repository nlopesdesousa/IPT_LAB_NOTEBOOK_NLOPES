```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks, savgol_filter

# --- Chargement des données ---
filename = "AA16(1).csv"
data = np.loadtxt(filename, delimiter=',')
Time = data[:, 0]
Force = data[:, 1]

# --- Lissage pour réduire le bruit ---
Force_smooth = savgol_filter(Force, 51, 3)

# --- Détection automatique des grands pics d'oscillation ---
# On fixe une prominence adaptative
prominence = 0.05 * np.max(Force_smooth)
distance = len(Time) // 200   # évite de détecter plusieurs points dans un même pic

peaks, _ = find_peaks(Force_smooth, prominence=prominence, distance=distance)

# --- Extraction des coordonnées ---
peak_times = Time[peaks]
peak_values = Force[peaks]

# --- Impression des résultats ---
print("\n--- Pics détectés ---")
for i, (t, y) in enumerate(zip(peak_times, peak_values), 1):
    print(f"Pic {i:02d} : t = {t:.5f} s   Force = {y:.3f}")

# --- Tracé pour vérification ---
plt.figure(figsize=(15,6))
plt.plot(Time, Force, color='skyblue', label='Signal brut')
plt.plot(Time, Force_smooth, color='blue', linewidth=1.2, label='Signal lissé')
plt.plot(Time[peaks], Force_smooth[peaks], 'ro', label='Pics détectés')
plt.title("Détection automatique des pics d’oscillations amorties")
plt.xlabel("Temps (s)")
plt.ylabel("Force (N)")
plt.xlim(0.5, 1.5)
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# === Lecture du fichier Excel ===
# Remplace par ton chemin
df = pd.read_excel(r"/home/IPT/IPT Lab Notebook/Kinetic_voltmeter/17-10-25/rebond_mesure/donneegrapheh.xlsx", header=[0,1])

# La première ligne contient les capacités (par exemple 1349, 1200, 1300)
capacites = df.columns.levels[0]
print(capacites)

plt.figure(figsize=(8,6))
h_max1 = []
dh_max1 = []
h_max2 = []
dh_max2 = []
for cap in capacites:
    bloc = df[cap].dropna()
    n = bloc["rebond"]
    h1 = bloc["h_n1"]
    dh1 = bloc["dh_n1"]
    h_max1.append(1e2*h1[0])
    dh_max1.append(1e2*dh1[0])
    h2 = bloc["h_n2"]
    dh2 = bloc["dh_n2"]
    h_max2.append(1e2*h2[0])
    dh_max2.append(1e2*dh2[0])


plt.grid()
plt.scatter(capacites, h_max1, label='Experiment 1', marker='o')
plt.errorbar(capacites, h_max1, yerr=dh_max1, fmt='o', capsize=2)
plt.scatter(capacites, h_max2, label='Experiment 2', marker='s', color='tab:red')
plt.errorbar(capacites, h_max2, yerr=dh_max2, fmt='s', capsize=2, c='tab:red')
plt.xlabel('Passed Capacity [mAh]')
plt.ylabel('First Bounce Height [cm]')
plt.title('First Bounce Height as function of Passed Capacity')
plt.legend()
plt.savefig("FirstBounceComparison.png")
```

```python
# Sélection du bloc correspondant à la capacité 1300
# cap est la liste des capacités telles que lues dans le MultiIndex
h0 = 22.1
cap_1300 = capacites[-2]        # comme tu l'as demandé
bloc = df[cap_1300].dropna()

n = bloc["rebond"]
h1 = bloc["h_n1"]
dh1 = bloc["dh_n1"]
h_max1.append(1e2*h1[0])
dh_max1.append(1e2*dh1[0])
h2 = bloc["h_n2"]
dh2 = bloc["dh_n2"]
h_max2.append(1e2*h2[0])
dh_max2.append(1e2*dh2[0])


n = np.insert(n, 0, 0)        # insérer 0 au début de n
h1 = np.insert(h1, 0, h0)     # insérer h0 au début de h1
dh1 = np.insert(dh1, 0, 1e-6) 
h2 = np.insert(h2, 0, h0)     # insérer h0 au début de h1
dh2 = np.insert(dh2, 0, 1e-6) 

def compute_coeffs_and_errors(h, dh):
    if h.size < 2:
        return np.array([]), np.array([])  # pas assez de points
    # e_n pour n = 0..N-2 correspond aux rebonds 1..N-1
    e = np.sqrt(h[1:] / h[:-1])

    # propagation d'erreur :
    # delta e / e = 1/2 * sqrt( (dh_{n+1}/h_{n+1})^2 + (dh_n/h_n)^2 )
    frac1 = (dh[1:] / h[1:])  # évite division par 0 (on a nettoyé >0)
    frac0 = (dh[:-1] / h[:-1])
    rel_err = 0.5 * np.sqrt(frac1**2 + frac0**2)
    de = e * rel_err
    return e, de

coeffs1, err1 = compute_coeffs_and_errors(h1, dh1)
coeffs2, err2 = compute_coeffs_and_errors(h2, dh2)
    
# --- Graphe ---
plt.figure(figsize=(7,5))
#plt.plot(n[:-1], coeffs1, marker='o')
plt.errorbar(n[:-1]+1, coeffs1, yerr=err1, fmt='o-', capsize=3, label='Experiment 1', color='tab:red')
#plt.plot(n[:-1], coeffs2, marker='o')
plt.errorbar(n[:-1]+1, coeffs2, yerr=err2, fmt='s-', capsize=3, label='Experiemnt 2', color='tab:blue')
plt.axhline(1, 0, 7, linestyle='--', color='k',label='Elastic collision')
plt.grid()
plt.xlabel("Bounce number", fontsize=14)
plt.ylabel("Restitution coefficient", fontsize=14)
plt.title("Coefficient of restitution (1300 mAh capacity removed)", fontsize=16)
plt.legend(fontsize=14)
plt.savefig("RestitutionCoefficientEvolution.png")

plt.show()
```

```python
h0 = 22.1
cap = capacites[-2]        # comme tu l'as demandé
bloc = df[cap].dropna()

n = bloc["rebond"]
h1 = bloc["h_n1"]
dh1 = bloc["dh_n1"]
h_max1.append(1e2*h1[0])
dh_max1.append(1e2*dh1[0])
h2 = bloc["h_n2"]
dh2 = bloc["dh_n2"]
h_max2.append(1e2*h2[0])
dh_max2.append(1e2*dh2[0])


n = np.insert(n, 0, 0)        # insérer 0 au début de n
h1 = np.insert(h1, 0, h0)     # insérer h0 au début de h1
dh1 = np.insert(dh1, 0, 1e-6) 
h2 = np.insert(h2, 0, h0)     # insérer h0 au début de h1
dh2 = np.insert(dh2, 0, 1e-6) 


# ----------------------------------------------------
# PARAMÈTRES PHYSIQUES
# ----------------------------------------------------
m = 0.023  # masse en kg (exemple à remplacer)
g = 9.81   # gravité

# ----------------------------------------------------
# ÉNERGIE POTENTIELLE POUR CHAQUE REBOND
# ----------------------------------------------------

def compute_energy(h, dh, m=m, g=g):
    """
    Calcule l'énergie potentielle E_n = m g h_n
    et son incertitude dE_n = m g dh_n.
    """
    E = m * g * 1e-2* h          # énergie
    dE = m * g * 1e-2* dh        # incertitude associée
    return E, dE

# énergie avant chaque choc pour chaque expérience
E1, dE1 = compute_energy(h1, dh1)
E2, dE2 = compute_energy(h2, dh2)

# Exemple d'affichage
print("Énergies avant chaque choc (expérience 1) :")
for i in range(len(E1)):
    print(f"Rebond {i} : E = {E1[i]:.4f} J ± {dE1[i]:.4f} J")

print("\nÉnergies avant chaque choc (expérience 2) :")
for i in range(len(E2)):
    print(f"Rebond {i} : E = {E2[i]:.4f} J ± {dE2[i]:.4f} J")


dt = 0.004  # durée du "choc" ou de dissipation

def compute_power(E, dE, dt=dt):
    """
    Calcule P_n = E_n / dt et son incertitude.
    """
    P = E / dt
    dP = dE / dt
    return P, dP

P1, dP1 = compute_power(E1, dE1)
P2, dP2 = compute_power(E2, dE2)


# ----------------------------------------------------
# PLOT DE LA PUISSANCE + ZOOM À CÔTÉ
# ----------------------------------------------------

ig, ax1 = plt.subplots(figsize=(6,5))

# ----- Graphique principal -----
ax1.errorbar(n+1, P1, yerr=dP1, fmt='o-', capsize=3, label='Experiment 1', color='tab:red')
ax1.errorbar(n+1, P2, yerr=dP2, fmt='s-', capsize=3, label='Experiment 2', color='tab:blue')

ax1.grid()
ax1.set_xlabel("Bounce number", fontsize=14)
ax1.set_ylabel("Power [W]", fontsize=14)
ax1.set_title("Power before each bounce (1300 mAh removed capacity)", fontsize=16)
ax1.set_yscale('log')

ax1.legend()

plt.tight_layout()
plt.savefig("PowerEvolution1300_log.png", bbox_inches='tight')
plt.show()
```

```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

# Données
h0 = 22.5
bloc = df[capacites[-1]].dropna()
n = bloc["rebond"].values
h1 = 1e2 * bloc["h_n1"].values
dh1 = 1e2 * bloc["dh_n1"].values

n = np.insert(n, 0, 0)        # insérer 0 au début de n
h1 = np.insert(h1, 0, h0)     # insérer h0 au début de h1
dh1 = np.insert(dh1, 0, 1e-6) 

# Fonction à fitter : h(n) = h0 * E^(2n)
# on cherche k = ln(E)
def h_model(n, k):
    return h0 * np.exp(2.0 * k * n)

# Fit pondéré par les erreurs
popt, pcov = curve_fit(h_model, n, h1, sigma=dh1, absolute_sigma=True)

k_fit = popt[0]
E_fit = np.exp(k_fit)

# Erreur sur k et E
k_err = np.sqrt(pcov[0,0])
E_err = E_fit * k_err  # propagation d'erreur pour exp(k)

print(f"E ajusté = {E_fit:.4f} ± {E_err:.4f}")

plt.figure()
plt.errorbar(n, h1, yerr=dh1, fmt='o', capsize=2, label='Données')
n_fit = np.linspace(min(n), max(n), 200)
plt.plot(n_fit, h_model(n_fit, k_fit), 'r-', label=f'Fit : h0*E^(2n), E={E_fit:.3f}')
plt.xlabel('n')
plt.ylabel('h (cm)')
plt.grid()
plt.legend()
plt.show()    
```

```python
h0 = 22.1
bloc = df[capacites[-2]].dropna()
n = bloc["rebond"].values
h1 = 1e2 * bloc["h_n1"].values
dh1 = 1e2 * bloc["dh_n1"].values
h2 = 1e2 * bloc["h_n2"].values
dh2 = 1e2 * bloc["dh_n2"].values

# Insert initial point (rebound 0)
n = np.insert(n, 0, 0)
h1 = np.insert(h1, 0, h0)
dh1 = np.insert(dh1, 0, 1e-6)
h2 = np.insert(h2, 0, h0)
dh2 = np.insert(dh2, 0, 1e-6)

# ---- Compute restitution coefficients for experiment 1 ----
def restitution(h_i, h_ip1):
    return np.sqrt(h_ip1 / h_i)

e_01 = restitution(h1[0], h1[1])   # between rebounds 0 and 1
e_78 = restitution(h1[7], h1[8])   # between rebounds 7 and 8

# ---- Function to draw an arc between two rebounds ----
def draw_arc(ax, x0, x1, y0, y1, text, color='black'):
    xc = np.linspace(x0, x1, 50)
    mid = (y0 + y1)/2
    amp = 0.15 * max(y0, y1)  # arc amplitude
    yc = mid + amp * np.sin(np.linspace(0, np.pi, 50))
    ax.plot(xc, yc, color=color, linewidth=1.3)
    ax.text((x0+x1)/2, max(y0,y1) + amp*1.1, f"e = {text:.3f}",
            ha='center', va='bottom', fontsize=10, color=color)

plt.figure(figsize=(8,5))
ax = plt.gca()

# Main plots
ax.errorbar(n, h1, yerr=dh1, fmt='s', capsize=2, label='Experiment 1', c='tab:red')
ax.errorbar(n, h2, yerr=dh2, fmt='o', capsize=2, label='Experiment 2', c='tab:blue')

plt.xlabel('Bounce number')
plt.ylabel('Bounce Height [cm]')
plt.title(f'Height evolution ({capacites[-2]} mAh Passed Capacity)')
plt.grid()
plt.legend()

# ---- Draw arcs for rebounds 0→1 and 7→8 ----
draw_arc(ax, 0, 1, h1[0], h1[1], e_01, color='tab:green')
draw_arc(ax, 7, 8, h1[7], h1[8], e_78, color='tab:purple')

plt.tight_layout()
plt.savefig("heightevolution_with_arcs.png")
plt.show()
```

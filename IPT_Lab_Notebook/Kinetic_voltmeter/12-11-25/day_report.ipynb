First results of the data analysis for the rebound experiment (see [Rebound experiment analysis](rebond_mesure/analysedonnees.ipynb)) were done, showing expected behaviour : an increase of the bouncing property with the dischargement. For each battery, one can calculate the restitution coefficient $E$ (elasticity of a collision) by :

$$
E_n = \sqrt{\frac{h_{n+1}}{h_n}},
$$

where $n$ is the $n$-th bounce and $h_n$ its height. The height can be determined by :

$$
h_n = \frac{1}{2}g\Big(\frac{\Delta t_n}{2}\Big)^2,
$$

with $g = 9.81$ m/s$^2$ the gravity constant and $\Delta t_n$ the time length of the $n$-th bounce, measurable with the piezoelectric plate.
With the hypothesis that the restitution coefficient is constant across rebounds, one can establish the following expression for the $n$-th height :

$$
h_n = E^{2n}\cdot h_0,
$$

where $h_0$ is the initial height. By doing a with on the data, one can establish the coefficient $E$.

In parallel, reflections and ideas concerning the physical theoretical foundations of each experiment are proposed.

**Hypothesis :**

Owing to differences in the molecular structure of the anode, an alkaline battery is expected to damp mechanical vibrations to varying degrees, with greater damping for charged batteries. Based on the conclusions of the referenced source, the formation of percolation pathways through oxidation would enable more efficient transmission of vibrations, whereas in the absence of such pathways, vibrations are attenuated and dissipated by the mobility of zinc particles. 

The internal structure *may* also be modeled as a network of springs (representing bonds) with different stiffnesses, combined with dampers. However, this hypothesis may be overly simplistic, as the battery system and the reactions occurring during discharge are inherently complex.

**1.1** With a resonance box:\
Mechanical vibrations are transmitted to the battery and the resulting dissipation is measured. Mechanical vibrations are generated by exciting a tuning fork, which in turn excites a resonance box. The advantage of the tuning fork is that the mechanical vibrations are directly characterizable through the sound produced. By attaching the battery to the resonance box, the vibrational capacity of the system is modified, which is expected to affect the sound quality. This effect can be readily quantified, for example by measuring a damping factor. The physical quantity involved is likely the mechanical impedance, i.e., the ability of an object or material to resist mechanical vibrations.\
$\rightarrow$ This measurement approach is actually be overly optimistic : the vibration damping effect attributable to the battery is likely minimal, or even negligible, particularly if mechanical contact between the resonance box and the battery is not well ensured, when compared to the intrinsic damping of the resonance box itself.\
**1.2** Battery attached to the tuning fork :\
The general idea is the same, but the modified system is smaller in scale (battery + tuning fork). By attempting to excite vibrations in this modified system, we aim to identify differences in sound quality as a function of the battery type.\
$\rightarrow$ In pratice the mode of vibration of a tuning fork leads to making this idea succesful difficult, because adding the battery restrict à lot the vibrations and in consequences no sound is measurable, and no damping...


**2.$~$Piezoelectric sensor**\
**2.1** With two sensors at each end (vibration transmission) :\
By exciting one side of the battery with a mechanical vibration, we attempt to measure the vibration transmitted to the opposite terminal. This approach might be explain by a spring–mass model and the change in acoustic wave velocity as it propagates through a material. Different resonance frequencies are expected, reflecting differences in internal structure.\
**2.2** With a single sensor (mechanical impedance) :\
By applying a preload, the coupled piezo–battery system and its ability to vibrate are investigated. An alternating signal is applied to the system, and the resulting signal is measured; this response is expected to reflect the vibrational behavior of the system, since the electrical impedance of a piezoelectric element is related to the mechanical impedance of the coupled system.

**Hypothesis (for measuring the sound from a collision) :**

During a collision, part of the system’s energy can also be converted into sound. The hypothesis is that when an initially stationary battery is struck, the energy delivered by the impact is dissipated in the form of heat, sound, and through molecular bonds. If one assumes that molecular bonds dissipate less energy when the battery is discharged, it can be hypothesized that a larger fraction of the impact energy will be transmitted in the form of sound.

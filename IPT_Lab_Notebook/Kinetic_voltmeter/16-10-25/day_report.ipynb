#### Day report 16/10/25

A setup was implemented to test vibration transmission. The battery is fixed horizontally, with a pendulum positioned on one side so that its mass is in contact with the battery, while the opposite side is struck with a hammer.

<img src="vibrationtransmission.jpeg" width="500"/>

The initial results are not conclusive, as no visible difference is observed. Both configurations (charged and discharged battery) appear to transmit vibrations to the pendulum mass equally well. Possible explanations include: 
1) since both charged and discharged batteries share the same internal structure beyond the anode, this common structure may transmit vibrations effectively, and any additional damping in the charged battery may be minimal or too small to detect with this setup;
2) variations may exist, but the setup lacks sufficient precision. In particular, the force applied by the hammer cannot currently be controlled, making exact comparisons difficult, and the battery support is suspended (no clamp attached to the table was used), so the battery cannot be fully immobilized. As a result, when the impact is strong, the entire support may move along with the battery, which could explain the similar observed behavior, dominating over the transmission within the battery itself.

Setup Improvements / Additional Tests: \
To properly assess the feasibility of the experiment, the setup needs to be improved. Successive tests, striking either the base (green) or the retaining ring, show that the pendulum can be excited without directly contacting the battery. Several trials were conducted applying a downward force on the support to increase stability and striking with small force, with slow-motion video recorded for later analysis; however, no clear differences are currently visible. Another challenge is determining the energy transmitted to the pendulum. This would require either an accelerometer or video analysis with a reference scale to measure maximum height. However, smaller applied forces (to avoid disturbing the system and affecting results) produce smaller pendulum amplitudes, highlighting the need for a setup that allows higher forces without introducing pendulum excitation through physical displacement of the apparatus. Additionally, measuring pendulum amplitude indirectly may accumulate too much uncertainty, whereas a piezoelectric sensor at the opposite end could provide a more precise measurement. \
From a theoretical perspective, it is important to clarify the mechanical impedance properties of the battery if this approach is pursued (i.e., its “resistance to vibration”).

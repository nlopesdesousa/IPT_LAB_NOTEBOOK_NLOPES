#### Day report 29/10/25

Further thinking on the previous offset results (see day 20,22/10/25) :


Regarding the offset measurements: an apparent inconsistency was that the offset did not change with the clamping force, which is not intuitive. This can be explained by the fact that the initial clamping distance already pushes the battery beyond its linear elastic regime, meaning that additional clamping does not increase the restoring force and the battery simply absorbs the extra deformation. At the same time, the transmission of small sinusoidal excitations through the battery remains perfectly linear, which is not contradictory since the dynamic deformations are on the nanometer scale, whereas variations in clamping distance are on the micrometer scale. In this regime, the system can be locally linearized, yielding an effective stiffness $k_{\text{eff}}$ around a given clamping distance, from which a local Young’s modulus can be derived in two equivalent ways for a given excitation frequency. This in principle allows extraction of an effective density per battery, which could be compared with a suitable internal model.

However, this approach requires accurate phase measurements, which are not achievable with the current oscilloscope, whose precision is limited to a few degrees—much larger than the inter-battery variations. Possible alternatives would involve using a significantly more precise phase-measurement instrument or increasing the effective propagation length (e.g., by placing several batteries in series), but these solutions are complex and may still fail to produce measurable phase differences, even when sweeping the excitation frequency from the hertz to the megahertz range. As a result, unless a reliable method for determining propagation velocity can be identified, this setup is unlikely to yield quantitative elastic parameters. The offset–battery correlation remains robust and easy to measure, but the offset itself appears to be only an indirect marker of stiffness differences and not directly exploitable for determining material properties.

Revisiting the previous data analysis of the rebound experiment :\
For the initial data analysis, the restitution coefficient was determined by averaging its values over all rebounds for a given battery, using an exponential fit. However, this approach implicitly assumes that the restitution coefficient remains constant across successive rebounds, which is not the case, which make more sense to see the increasing of the bouncing proprety on only the coefficient of restitution (ratio of kinetic energy) only for the first rebound : 

<img src="FirstCoefficientRestitutionComparison.png" width="600"/><img src="RestitutionCoefficientEvolution.png" width="600"/>

Indeed, it is observed that the restitution coefficient increases over successive rebounds, indicating that the collisions become progressively more elastic as the number of impacts increases. This observation motivated the calculation of the evolution of the power just before each rebound.

<img src="PowerEvolution700_log.PNG" width="360"/><img src="PowerEvolution1000_log.PNG" width="360"/><img src="PowerEvolution1300_log.png" width="360"/>

It can therefore be inferred that two distinct energy regimes can be identified. The first rebound occurs in a significantly higher energy domain compared to the subsequent rebounds, which may account for the observed behavior of the restitution coefficient. This can be interpreted as follows: at lower energies, damping associated with the displacement of zinc particles no longer plays a significant role, and the system consequently behaves like a “normal” elastic solid during rebound.

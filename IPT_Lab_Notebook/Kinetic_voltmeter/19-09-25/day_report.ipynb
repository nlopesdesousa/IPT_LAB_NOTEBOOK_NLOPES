10h : Discussion and identification of the key points of the problem : \
$\rightarrow$ We want to explain the bouncing phenomenon observed when a battery is discharged. \
Hypothesis: A battery supplies energy through an oxidation reaction (release of an electron and formation of a zinc oxide molecule) that changes the molecular bonds. Before discharge, the bonds are not aligned and dampen the impact energy, whereas after discharge, the bonds are aligned and transmit the energy.
\
\
Two facets of the problem :
1. Establish the relationship between the state of charge and the bouncing of a battery (show that there is indeed a correlation between these two aspects) and potentially find a way to evaluate the reliability of the test.
2. Justify the fact that the oxidation reaction induces changes in molecular bonding, which would reduce internal friction and therefore result in less damping.

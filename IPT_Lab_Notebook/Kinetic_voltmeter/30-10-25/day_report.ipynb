##### Reflexion on the influence of clamping distance/mecanism on the system

The piezoelectric sensor measures time-dependent variations of mechanical stress, because in real situation the sensor is connected to a finite impedance (oscilloscope). The tension is then 

$$
V(t) \propto \frac{d\epsilon(t)}{dt},
$$
with $\epsilon$ the deformation. The application of an additional static compression due to clamping produces only a transient electrical response, after which the signal stabilizes. Once a steady state is reached, a constant applied stress does not affect the measured signal.
Consequently, the clamping force, provided it remains constant over time, has no influence on the recorded signal. This conclusion allowed the experiment to be conducted with the setup oriented vertically, as the imposed deformation was constant and did not impact the measurement.
This configuration offered practical advantages, notably easier battery replacement and the possibility of applying only minimal clamping, sufficient to ensure mechanical contact between the piezoelectric element and the battery surfaces. Measurements performed under this configuration yielded results consistent with those obtained previously.

##### Reflexion on the stiffness model and the cause of the offset

Additional tests were performed using samples other than batteries in order to identify another object that might exhibit similar behavior in terms of density. Experiments were conducted using materials that appeared to be aluminum (easily deformable) and iron. In these cases, no measurable offset was observed, which was inconsistent with our initial hypothesis based on modeling the system as springs with different stiffnesses.

This led to the realization that the main difference between the two experiments (with metallic samples versus batteries) is that batteries are electrically charged and exhibit a potential difference between their terminals. Charged batteries present a voltage of approximately 1.5 V, while discharged batteries exhibit voltages around 1.2 V, values that are strikingly close to the offsets measured by our experimental setup.

Although the exact mechanism by which the piezoelectric sensor detects this voltage is not fully understood, it is likely that, since the surface of the piezoelectric element is in direct contact with the battery, the sensor effectively measures the electrical potential difference. Consequently, the experiment does not purely probe a mechanical effect but rather behaves as a voltage measurement, similar to a voltmeter, and therefore constitutes an electrical rather than a mechanical method.

Sketch : Setup idea to measure the mechanical impedance

The idea is to establish purely mechanical contact between the battery and the piezoelectric sensor, while electrically isolating the contact surface, and to measure the sensor response within an electrical circuit by monitoring the current. The measured current is related to the excitation signal through
$$
V_i = Z\cdot I(t),
$$
where $Z$ denotes the electrical impedance of the circuit. This impedance is assumed—subject to further verification—to be related to the mechanical impedance (i.e. resistance to vibration), which may be influenced by the internal structure of the battery and its ability to damp mechanical vibrations. To measure $I(t)$, the alternating voltage across a known resistance is recorded.

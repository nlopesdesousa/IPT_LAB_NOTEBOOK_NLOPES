Analysing them by associating it to an damped forced oscillator andd computing the $Q$ factor (inverse of the damping coefficient) :

$$
m\ddot{x}(t) + c\dot{x} + kx(t) = F_0\sin(\omega t) ~~~~~ Q = \frac{1}{2\zeta}, ~~~~\zeta = \text{damping coefficient}
$$

Different trends of the $Q$ factor were observed, in particular indicating that the more discharged the battery is, the stronger its damping effect. These measurements were performed in a low-energy regime, since the spectrum analyzer is limited to output voltages of only a few hundred millivolts. Nevertheless, this result is significant, as it suggests that charged and discharged batteries may already be distinguishable in a low-energy domain.

The next step is to confirm the reproducibility of these trends. If validated, this would already provide a partial answer to the initial research question. To physically interpret and justify these observations, the gain curves could be decomposed into a set of harmonic oscillators associated with the dominant resonance frequencies. By analyzing, for instance, which resonant frequencies exhibit systematic variations in their quality factor, one can formulate hypotheses about their physical origin within the system. Resonances showing no clear trend could be attributed to the battery casing or to the experimental setup itself, with the casing likely corresponding to the dominant structural resonance, as it constitutes the largest fraction of the vibrating mass. Additional resonances may be artifacts related to the setup, such as the metallic clamping mechanism. In contrast, resonances exhibiting clear trends can reasonably be associated with internal changes in the battery.

Complementary global indicators could also be considered, such as the mean gain, which would reflect the overall transmission efficiency, or the standard deviation of the gain, where larger values would indicate sharper and higher resonance peaks and could be interpreted as a generalized quality factor. A plausible explanation for the observed low-energy behavior is a change in the effective viscosity of the gel (excluding the zinc grains), which may be higher in discharged batteries than in charged ones. This would naturally lead to lower quality factors for discharged batteries, in agreement with the structural differences observed during battery dissection.

The remaining challenge is to access the appropriate energy regime. To this end, and while continuing to use the spectrum analyzer, an amplifier could be placed at its output in order to inject a larger amount of energy into the piezoelectric actuator. A second attenuation or conditioning stage at the output would then ensure that the signal remains within the acceptable input range of the spectrum analyzer, allowing the damping effects associated with the zinc grains to be probed.

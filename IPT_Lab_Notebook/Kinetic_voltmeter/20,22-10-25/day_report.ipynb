#### Day report 20,22/10/25

Synthesis of tests-results-conclusions of several different experimentations : \
With the same setup from day 17/10/2025, tests showed no observable change in gain when varying the signal amplitude from 0 to 15 V (the maximum allowable voltage for the piezo elements). \
Likewise, the phase did not exhibit any systematic variation that could be used with our equipment to identify differences in propagation speed. \
It is also important to ensure correct wiring of the piezo elements; the polarity must not be reversed. A simple way to verify this is to check that the measured gain is greater than 1.

Somewhat by chance, when removing and reinserting the batteries, it was observed (in AC mode) that each battery produced a distinct and reproducible phase difference. It later became clear that the measurement procedure was incorrect: the phase difference was estimated under the assumption that the two signals would not shift vertically. The signals were aligned once, and all subsequent measurements were performed relative to that alignment. This method does not measure a true phase shift, because inserting different batteries causes the received signal to undergo a vertical shift relative to the driving signal. Consequently, the measurement is not physically meaningful.
However, the fact that the results remained reproducible indicated the presence of a real underlying effect.

This effect was identified as a reproducible variation of the DC offset of the receiving piezo depending on the battery. In static conditions, this can be explained by the fact that, when the battery is clamped between the piezos, the receiving piezo has a fixed stiffness while the battery has a stiffness that depends on its state of charge. The equilibrium contact point between the receiving piezo and the battery therefore differs from one battery to another. In other words, the equilibrium indentation of the receiving piezo depends on the specific battery, leading to a reproducible offset.

An attempt was made to measure this equilibrium voltage directly by placing the battery between the piezos without excitation and reading the voltage with a standard voltmeter. This method fails because the piezo charge leaks too quickly (the voltmeter input impedance is too low), causing the measured value to drop too rapidly for a reliable reading.

A reliable measurement of the piezo equilibrium voltage can instead be obtained by driving the excitation piezo with a sinusoidal signal. In this configuration, the offset of the receiving piezo corresponds to the mean value of its oscillation, which directly reflects the equilibrium voltage. Because the oscillation frequency is high, charge leakage is negligible over a single period, and the measured offset is stable and reproducible.

A major advantage of this setup is that the clamping force has no influence on the measurement. By imposing a prescribed sinusoidal motion on the excitation piezo, its behaviour becomes independent of its pre-stress. As long as the excitation frequency and amplitude are kept constant and the piezo operates within its linear regime, the measurement is fully reproducible.

The offset of the receiving piezo can, in principle, be converted into an equivalent mechanical displacement. A simple force-balance analysis should then allow this displacement to be related to the stiffness of the battery (assuming the stiffness of the piezo is known). This would make it possible to construct a curve of the effective battery stiffness as a function of the extracted capacity.


With favourable conditions, one could then attempt to model the battery stiffness based on its chemical state, compare the model with the experimental results, and potentially achieve good agreement. (we have to be optimistic)

All of this raises questions regarding the previous results. A variation in the effective spring constant implies a change in the effective Young’s modulus, and therefore a change in both the resonance frequency and the propagation velocity. Since it was previously observed that the clamping force shifts the resonance frequency, it would be worthwhile to repeat the experiment with the spectrum analyzer while minimizing variability in the clamping conditions. It is also possible that a measurable damping effect may appear. Similarly, the propagation velocity should also exhibit noticeable changes; this remains to be investigated.

Upon further reflection, a well-defined resonance frequency cannot truly be extracted in this configuration, because the boundary conditions are imposed sinusoidally. However, some calculations lead to an interesting expression for the effective Young’s modulus (which should, in principle, be directly related to the spring constant measurable with the piezos):

$$
\begin{equation}
 E_{eff} = \left(\frac{2\pi L}{f\phi}\right)^2\rho_{\text{eff}}
\end{equation}
$$

where $f$ is the excitation frequency (controlled experimentally), $L$ is the effective length of the battery, $\phi$ is the phase difference between the driving and received signals, and $\rho_{\text{eff}}$ is the effective density of the propagation medium. Because of the internal geometry of the battery, it is likely that the full battery length should not be considered; a more controlled clamping setup may help clarify this point. Preliminary observations suggest that the phase does not vary significantly under different clamping conditions.

If the measurements behave favourably (a highly optimistic scenario), it should be possible to compute $E_{\text{eff}}$ and then infer the corresponding spring constant. This would allow a comparison between two quantities:
1. the experimentally determined spring constant from the indirect piezoelectric measurement, $k_{\text{exp}}$, and
2. the spring constant derived from an internal mechanical model of the battery, $k_{\text{mod}}$.

This remains speculative, but it provides a potential path toward correlating mechanical measurements with internal battery properties.

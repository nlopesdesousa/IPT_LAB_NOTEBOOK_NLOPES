```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks, savgol_filter

# --- Chargement des données ---
filename = "AA2(2).csv"
data = np.loadtxt(filename, delimiter=',')
Time = data[:, 0]
Force = data[:, 1]

# --- Lissage pour réduire le bruit ---
Force_smooth = savgol_filter(Force, 51, 3)

# --- Détection automatique des grands pics d'oscillation ---
# On fixe une prominence adaptative
prominence = 0.05 * np.max(Force_smooth)
distance = len(Time) // 200   # évite de détecter plusieurs points dans un même pic

peaks, _ = find_peaks(Force_smooth, prominence=prominence, distance=distance)

# --- Extraction des coordonnées ---
peak_times = Time[peaks]
peak_values = Force[peaks]

# --- Impression des résultats ---
print("\n--- Pics détectés ---")
for i, (t, y) in enumerate(zip(peak_times, peak_values), 1):
    print(f"Pic {i:02d} : t = {t:.5f} s   Force = {y:.3f}")

# --- Tracé pour vérification ---
plt.figure(figsize=(15,6))
plt.plot(Time, Force, color='skyblue', label='Signal brut')
plt.plot(Time, Force_smooth, color='blue', linewidth=1.2, label='Signal lissé')
plt.plot(Time[peaks], Force_smooth[peaks], 'ro', label='Pics détectés')
plt.title("Détection automatique des pics d’oscillations amorties")
plt.xlabel("Temps (s)")
plt.ylabel("Force (N)")
plt.xlim(0.45, 0.65)
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# === Fonction de fit exponentiel ===
def expo_decay(n, h0, k):
    return h0 * np.exp(-k * n)

# === Lecture du fichier Excel ===
# Remplace par ton chemin
df = pd.read_excel(r"/home/IPT/IPT Lab Notebook/Kinetic_voltmeter/19-09-25/rebond mesure/donneegrapheh.xlsx", header=[0,1])

# La première ligne contient les capacités (par exemple 1349, 1200, 1300)
capacites = df.columns.levels[0]

# === Graphe Expérience 1 ===
plt.figure(figsize=(8,6))
for cap in capacites:
    bloc = df[cap].dropna()
    n = bloc["rebond"]
    h = bloc["h_n1"]
    plt.scatter(n, h, label=f"{cap} mAh")
    try:
        popt, _ = curve_fit(expo_decay, n, h, p0=(h.iloc[0], 0.5))
        n_fit = np.linspace(1, n.max(), 100)
        plt.plot(n_fit, expo_decay(n_fit, *popt), '--')
    except:
        pass

plt.title("Évolution de la hauteur (Expérience 1)")
plt.xlabel("Nombre de rebond n")
plt.ylabel("Hauteur hₙ (m)")
plt.legend(title="Capacité (mAh)")
plt.grid(True)
plt.tight_layout()
plt.show()

# === Graphe Expérience 2 ===
plt.figure(figsize=(8,6))
for cap in capacites:
    bloc = df[cap].dropna()
    n = bloc["rebond"]
    h = bloc["h_n2"]
    plt.scatter(n, h, label=f"{cap} mAh")
    try:
        popt, _ = curve_fit(expo_decay, n, h, p0=(h.iloc[0], 0.5))
        n_fit = np.linspace(1, n.max(), 100)
        plt.plot(n_fit, expo_decay(n_fit, *popt), '--')
    except:
        pass

plt.title("Évolution de la hauteur (Expérience 2)")
plt.xlabel("Nombre de rebond n")
plt.ylabel("Hauteur hₙ (m)")
plt.legend(title="Capacité (mAh)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

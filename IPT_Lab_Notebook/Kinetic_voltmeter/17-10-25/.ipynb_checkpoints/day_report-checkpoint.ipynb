#### Day report 17/10/25

8h : The day began with a discussion on the day’s program, sharing personal reflections, and re-evaluating the experiments and the problem. Key points included :
1. In theory, internal changes in the battery should affect its moment of inertia. It was considered whether this property could still be exploited, for example by using a torsion pendulum to measure the period and compare moments of inertia. Several challenges were noted, including how to measure the period (e.g., laser?), determine the pendulum’s restoring constant, find an object with the same mass, shape, and known moment of inertia, attach the battery, and whether the expected difference would be measurable given limited experimental precision.
2. Reflection on the 16/10/25 setup, with proposed improvements: securing the battery more rigidly, adding a rotating system with a metal rod, and conducting experiments using a piezoelectric sensor at one end while controlling initial energy with a pendulum on the other end.
3. Exploiting vibrational response using a piezoelectric sensor for different tests.

9h : A piezoelectric setup was installed using a function generator to excite one piezo and an oscilloscope to record the response on the other. The battery was clamped between two walls, with a piezoelectric sensor placed at each end (between the battery and the wall). The clamping distance was set to 29.45 mm, and a frequency sweep was performed. Amplitudes and corresponding frequencies were recorded for three minima and three maxima. The experiment was repeated by loosening and re-tightening the same battery to the same clamping distance, with extrema measured again to estimate variability within a single experiment.

10h : Continuation of the previous experiment with charged battery, the previous experiment was continued using a charged battery. The general behavior was the same, with similar extrema and amplitudes, raising questions about the setup. A hypothesis was proposed that the battery may be clamped too tightly between the walls, causing amplitude saturation or excessive constraint, which could prevent the internal behavior from being revealed. \
11h : Setup modification with foam spacers : Foam pieces were acquired to add thickness at the ends of the battery, either on one side or both, allowing more freedom of movement. The experiment was repeated with foam added at the non-excited end of the battery. Additional tests were conducted with the modified setup, and it was decided to also display the input signal applied to the piezoelectric sensor.

Observations : The previously measured general behavior (minima and maxima) was not specific to the battery but rather related to the sensor. It was observed that the input signal displayed on the oscilloscope did not match the expected settings, an effect that disappeared when the cable connected to the sensor was disconnected. This indicates that the output signal is altered by the connection to the sensor, either because the sensor draws too much current, preventing the expected amplitudes, or because reflected waves impact the exciting piezo, creating interference between the generator’s excitation voltage and the returned excitation. The second possibility is considered less likely, as the excitation signal frequency observed on the oscilloscope is preserved, whereas a phase shift would otherwise modify it. This implies that the previously observed amplitude variations are largely influenced by this sensor-related behavior. In any case, since the excitation frequencies are preserved, the issue is likely due to the current drawn by the sensor, and the experiment can be repeated using the gain instead.

13h-15h : Experiment Repetition Using Gain Analysis, the experiment was repeated, this time analyzing the gain. Frequencies from 100 kHz to 1 MHz were measured in 100 kHz increments, with finer measurements around the observed resonance peak. This was performed for both a fully charged and a fully discharged battery. However, a second confirmation measurement on one of the batteries showed that the variability of the experiment is greater than the precision required to detect differences. Several key observations were made : 
1. Over the 100 kHz – 1 MHz sweep, the resonance curves appear visually identical at first glance. However, with repeated measurements, it is possible that the quality factors ($Q$) of certain resonance peaks vary slightly between charged and discharged batteries.
2. The resonance peak frequencies appear to shift depending on the clamping force of the setup. This suggests that the variability of manually obtained resonance curves is entirely determined by uncontrolled variations in the setup. Even with controlled clamping force, it seems impossible to improve reproducibility with this setup. Additionally, foam spacers do not appear beneficial; tests with thicker foam indicated even greater sensitivity to clamping. This should be confirmed with tests without foam, but it seems the foam may not provide any advantage.
3. An additional test involved placing the battery upright on a piezo sensor to observe visible resonances; none were apparent during the experiment, possibly due to incorrect implementation.

Upon reflection, it may be expected that no results are visible with these experiments. Although even a minimal change in geometry should slightly influence the observed resonance frequencies, the current experiments test the battery’s response to a continuous stimulus, whereas the only successful experiment (the rebound test) observes the response to a brief, intense stimulus. Two experimental approaches seem possible while retaining this setup :
1. Introduce an impulse to the exciting piezo and analyze the input and output FFTs to examine propagation speeds and related parameters. Alternatively, one could maintain continuous excitation at the resonance frequency and measure the phase shift between input and output signals to infer propagation speed. Since gradual geometric changes do not appear to significantly affect resonance frequencies, these methods would directly exploit the gradual changes in stiffness and composition of the anode and cathode, which should directly impact the propagation speed of mechanical waves.
2. To simulate the intensity of a rebound, the amplitude of the excitation signal could be increased. This might reveal a noticeable difference in gain at the resonance frequency. In a fully charged battery, the anode consists of zinc particles suspended in a gel, which may have too high a viscosity to produce measurable damping under small excitations, potentially explaining the lack of observed results so far. By controlling the signal amplitude more precisely, it may be possible to reach a damping threshold at which the zinc particles begin to move, producing visible and quantifiable damping differences between batteries.

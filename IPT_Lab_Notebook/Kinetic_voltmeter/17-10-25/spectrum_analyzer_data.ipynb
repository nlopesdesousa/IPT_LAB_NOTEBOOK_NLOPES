```python
import pandas as pd
import matplotlib.pyplot as plt

#  Remplace le nom du fichier par le tien :
filename = "P1.TXT"

#  Lire le fichier en ignorant les lignes de mtadonnes au dbut
# On dtecte la ligne qui commence par "Frequency" pour sauter ce quil y a avant
with open(filename, 'r') as f:
    lines = f.readlines()

# Trouver la ligne o commence le tableau
for i, line in enumerate(lines):
    if line.strip().startswith('"Frequency"'):
        start = i
        break

# Charger les donnes  partir de cette ligne
df = pd.read_csv(filename, 
                 sep=r'\s+',            # sparateur = espaces ou tabulations
                 skiprows=start,        # saute les lignes den-tte
                 quotechar='"')         # pour grer les guillemets

# Vrification
print(df.head())

#  Extraire les colonnes d'intrt
x = df["Frequency"]
y = df["Data Trace Real"]

#  Tracer
plt.figure(figsize=(8,5))
plt.plot(x, y, color='blue', lw=1)
plt.title("Rponse en amplitude (Log Mag) du canal 1")
plt.xlabel("Frquence (Hz)")
plt.ylabel("Amplitude (dB)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

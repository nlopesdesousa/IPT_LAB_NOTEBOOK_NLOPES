```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile

def analyse_frequence_audio(fichier_audio):
    # Lire le fichier audio
    taux_echantillonnage, donnees = wavfile.read(fichier_audio)

    # Si le fichier est stéréo, on ne garde qu'un seul canal
    if donnees.ndim > 1:
        donnees = donnees[:, 0]

    # Nombre total d'échantillons
    N = len(donnees)

    # Appliquer la FFT
    fft_result = np.fft.fft(donnees)
    freqs = np.fft.fftfreq(N, d=1/taux_echantillonnage)

    # Garder seulement les fréquences positives
    freqs_pos = freqs[:N // 2]
    fft_magnitude = np.abs(fft_result[:N // 2])

    # Trouver la fréquence minimale et maximale avec une magnitude significative
    seuil = max(fft_magnitude) * 0.01  # seuil de 1% de l'amplitude max
    indices_utiles = np.where(fft_magnitude > seuil)[0]

    if len(indices_utiles) == 0:
        print("Aucune fréquence significative détectée.")
        return

    freq_min = freqs_pos[indices_utiles[0]]
    freq_max = freqs_pos[indices_utiles[-1]]

    print(f"Plage de fréquence détectée : {freq_min:.2f} Hz à {freq_max:.2f} Hz")

    # Affichage du spectre
    plt.figure(figsize=(10, 6))
    plt.plot(freqs_pos, fft_magnitude)
    plt.title("Spectre de fréquences (FFT)")
    plt.xlabel("Fréquence (Hz)")
    plt.ylabel("Amplitude")
    plt.grid(True)
    plt.xlim(0, taux_echantillonnage / 2)
    plt.show()

# Exemple d'utilisation
fichier = "neuve1choc.wav"  # Remplace par le chemin de ton fichier
analyse_frequence_audio(fichier)
fichier2 = "neuve2choc.wav"  # Remplace par le chemin de ton fichier
analyse_frequence_audio(fichier2)
fichier3 = "vieille1choc.wav"  # Remplace par le chemin de ton fichier
analyse_frequence_audio(fichier3)
fichier4 = "vieille2choc.wav"  # Remplace par le chemin de ton fichier
analyse_frequence_audio(fichier4)
```

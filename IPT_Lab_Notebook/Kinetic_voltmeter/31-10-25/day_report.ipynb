The spectrum analyzer was then used to determine the resonance frequencies and to check for differences in amplitude. The results were inconclusive, as no spectral differences were observed. Given that the spectrum is identical for the piezo alone and with a battery placed on it, both with and without pre-compression, it can be inferred that the piezo’s impedance is much higher than the contribution of the battery to the piezo–battery system. Furthermore, it is likely that the resonance frequencies in the setup lie outside the range accessible to the spectrum analyzer, making measurements impossible. In this context, a configuration with two piezos under compression produces a visible resonance peak in the analyzer spectrum, which could then be used to interpret the battery’s mechanical impedance as the ratio between the input force and the transmitted force.\
More generally, no difference was observed between batteries across the various piezo experiments...

**Hypothesis**\
The main hypothesis is that there exists a threshold above which damping becomes observable. Indeed, the probable cause of damping has previously been attributed to the movement of zinc particles within the anode gel. As the battery discharges, more of these zinc particles are converted into zinc oxide, forming a rigid structure, which implies that damping decreases as the battery discharges. Since the gel is viscous, there is a minimum force required for the zinc grains to move freely, corresponding to an energy threshold. It is therefore likely that the piezo was unable to supply sufficient energy. In particular, the spectrum analyzer’s output power was limited to a maximum of 20 dBc, already below the 15 V maximum supported by the piezos. Even when applying 15 V with a signal generator, no amplitude damping could be observed. Consequently, it would be necessary to determine the minimum energy required to produce a visible (or measurable) damping effect.

To exhaust the list of feasible experiments, we propose these three approaches :
1. Use piezos capable of larger displacement and greater force, and repeat the previous experiments.
2. Investigate damping using an alternative method. Attach a DC motor to rotate the battery around its vertical axis. By applying a high voltage to the motor (corresponding to high rotational speed), we wait until the rotation stabilizes, then cut the generator voltage. The voltage generated by the residual motion and its decay is then measured. A faster decay would be expected if the battery exhibits higher damping. However, under the threshold hypothesis, differences in decay rate would likely only be observable at the beginning, when the speed is sufficiently high. Another consideration is that, if observable damping is indeed caused by the movement of zinc particles, centrifugal forces would concentrate these particles at the periphery, potentially reducing measurability over time. A further variant of this experiment would be to apply an AC voltage to the generator continuously, without ever cutting it. Due to the battery’s inertia, the motor may not follow the applied voltage instantaneously, and this effect could be more or less pronounced depending on the internal damping of the battery. Changes in the battery’s moment of inertia could also be considered, though intuitively this effect is negligible, as rebound phenomena are primarily observed through energy transfer.

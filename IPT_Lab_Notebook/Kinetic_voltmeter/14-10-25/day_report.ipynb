Observations+conclusions: 
1. just by sticking tape on one of the branch, the sound is significantly reduced. This can easily be explain by the fact that the tape is constraining the mouvement of the branch, which results in less sound. This means that already a lot of the sound is reduced by just sticking tape, which implies on one side that adding the battery (way heavier) will probably reduced the noise a lot more which might be very difficult to measure, and on the other side it means that we can not attribute the damping of the sound solely to the battery, because the tape itself already has an impact on it.
2. the experiment which adds the battery on of the branch confirmed some of the worries : the sound is completely attenuated, the branch isnt able to vibrate. It has been however observed that by sticking the battery lower on the branch the sound was more noticeable, but unfortunately not loud enough to measure appropriately.
3. By testing different positions of the battery on the branch the following conclusions were done : the sound greatly depends on the way the battery is stuck. If tightly stuck the branche cannot vibrate because the mouvement is completely constrained by the system battery + tape. It it is less tightly stuck, the branch can move a bit more freely, which enable some sound to be created but it is quickly damped by the impact of the surface of the battery with the one of the branch. It is then very difficult with the present set up to evaluate correctly and reproduced exactly the same way to hang the battery on the branch. Moreover the measure of a damping of the sound with this set up probably significantly depend tree phenomenon : a) the tape used to maintain the battery constraining already itself the mouvement, b) the weight, shape and height of the battery on the branch, which contrain the vibration, c) the small impact of the battery and the branch, which actually would use the same phenomenon as the bouncing test, using the property to bounce or damp.

All of this obeservations and conclusions lead to the fact that this experiment, as it is, is not good enough to establish a correspondance between the internal structure of the battery (that we expect to be able to more or less damp acoustic or mecanic waves) and the sound created by vibration of a tunning fork.

3. Using the pendulum setup again, the battery was given an initial rotation by twisting it three times around itself to study its behavior. Videos were recorded for later analysis, but no macroscopically distinguishable phenomena were observed. The idea was to use the possible change in inertia of the battery. The set up was really imprecise, the initial angle and the height were not particularly consistant. Because the internal mass repartition must be different this track could show some results. But measuring such a difference — which we believe to be small, since the order of magnitude of these internal differences is small (the diameter of a battery is about 2 cm) — would require very high precision. However, for such a setup, the measuring tools at our disposal are probably not accurate enough. Moreover, evaluating a torsion pendulum would also require knowing the restoring constant of the wire, which is not easily determined.

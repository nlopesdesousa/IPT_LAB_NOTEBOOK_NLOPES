#### Day report 14/10/25

Series of small tests and conclusions of the day :
1. Return to the setup with the battery taped to the tuning fork, this time using paper tape. A strong effect on the sound was observed just by sticking the tape onto the edges of the branch (without the battery). Several trials were conducted, but it was impossible to reproduce the behavior observed during the very first experiment. I believe that the tape itself — and the contact surface with the tuning fork — has a major influence on the results.

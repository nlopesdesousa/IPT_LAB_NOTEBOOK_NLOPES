Regarding the piezoelectric measurements, we focused on confirming the observed trends using only the old piezos. We observed a decrease in the quality factor with an approximately linear trend, analyzing primarily the main peak using

$$
Q=\frac{f_{res}}{\Delta f},
$$
with $\Delta f$ the full width at half maximum of the peak. The experimental variation is relatively large, partly explained by cable resonances, as previously discussed. Another source of error may be related to mechanical clamping : the measured values strongly depend on the applied clamping force, and an undesired effect is that the quality factor—and thus damping—also depends on the clamping. When the setup is tightened excessively, an artificial damping is introduced, saturating the phenomenon and preventing any trend from being observed. To address this, the setup was given the minimum clamping required to obtain a measurable gain curve, allowing the trend to be observed while reducing uncertainty associated with clamping. The experiment was then repeated six times to achieve reasonable statistical significance, yielding the resulting gain curve.

Interestingly, we observe that the quality factor decreases with battery discharge (and thus damping increases), which is the opposite trend to that seen in the rebound experiments. This confirms the two-energy-level hypothesis. Indeed, the power required to reach the high-energy regime was previously estimated at 12 W. By calculating the maximum power of one of the old piezos, we find 

$$
P_{max} = Ff\delta = 1 \times 500 \cdot 10^{3} \times 100 \cdot 10^{-9} \approx 0.05 \text{W},
$$ 

placing the experiment clearly in the low-energy regime, where the zinc-grain damping effect cannot be observed.

The observed effect is opposite to that of zinc-grain damping. The approximately linear relationship can be explained by discharging the batteries at constant current, removing slices of capacity proportional to the progress of the redox reaction, which naturally produces a linear trend. 

<img src="RedOx.png" width="1000"/>

A qualitative explanation can be proposed for the observed decrease in quality factor with battery discharge (i.e., increased damping). In a fully charged battery, the anode consists primarily of gel and zinc grains, so at low power these grains cannot move relative to the gel, allowing waves to propagate uniformly through the anode. In a discharged battery, however, the anode is largely composed of a ZnO matrix enclosing zinc grains, with pockets of water between the percolation networks. In this case, waves propagate efficiently through the ZnO networks, but when encountering a water pocket, a lag is introduced, generating shear at the water–zinc interface and resulting in additional damping. It should be noted that this explanation is one hypothesis among several, and it does not appear feasible to verify it experimentally.

The [specific Notebook](../28-11-25/fftchoc.ipynb) is availaible in the folder "28-11-25". However, to identify potentially relevant features for analysis, we explored the use of machine learning. By employing various ML architectures (random forest, neural networks, etc.), we were able to develop a classification model capable of distinguishing between charged and discharged batteries. This model performed very well, achieving an accuracy of 92% and an F1 score of 92%. All the python code and data used for the machine learning are gathered in the folder "classification". More specifically the main code used is available [here](classification/battery_classifier_v5_corrected.ipynb). The analysis of the most important features used to train the model most effectively revealed that the MFCC and Contr coefficients are the most relevant. These are coefficients commonly used in speech recognition. MFCCs provide a characterization of the spectral shape, while Contr coefficients describe the spectral contrast, i.e., its roughness. A rough spectrum indicates many resonance peaks, suggesting that the material is more “resonant.” Using these features, we extracted the most informative characteristics related to differences observed during battery discharge. Since the model was trained on batteries from different brands and with strongly contrasting discharge levels (here quantified by the voltage a battery can deliver), we are reasonably confident that these coefficients reflect genuine discharge-related differences rather than brand-specific artifacts.

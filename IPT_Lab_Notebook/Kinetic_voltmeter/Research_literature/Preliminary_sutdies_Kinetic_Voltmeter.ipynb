##### Self study on battery itself (how it works, composition, etc.) :
Case of alkaline battery : storage of chemical energy (then transformed in electrical energy by chemical reaction that induce push force, repulsive Coulomb force, that allows flow of electrons).\
\
<img src="battery.png" width="500">\
\
essentially composed of a cathode (+) (mix manganese + graphite) and an anode (-) (mix powdered zinc + gel agent) electrically isolated. Electron flow possible through chemical reactions (oxydation - reduction) that either free electron that accumulate in the anonde (oxydation), or use electron for a reaction (reduction). A potential different is induced by accumulation of electron on - extremity that is measurable with voltmeter. The pushed force comes from repulsive coulomb force between electrons that will try to reach the point with less electron (cathode if circuit allows it). The chemical reaction runs if battery is in a circuit, but can be slowed down by the fact that the chemical reaction decreases with use. The lifetime depends on chemical energy quantity at the beginning, the energy quantity required by the charges in the circuit, the age and the temperature of the battery)
(video explaining how a alcaline battery works : https://www.youtube.com/watch?v=PXNKkcB0pI4)

##### Video sources of the phenomenon

1. « Battery Bounce Test - tell full from empty by dropping it! » https://www.youtube.com/watch?v=suH32X91sQE \
Several tests on different surface and brand of AA alkaline batteries to tell a full AA alkaline battery from an empty one just by dropping it. The full battery doesn't bounce, the empty one does bounce.

2. « Battery bounce test at idle – Why do they bounce when they are dead? » https://www.youtube.com/watch?v=V8Hro4Ud7-I \
Bouncing tests on different alkaline batteries, slow-motion video of the fall and the resulting bouncing. Some battery bounces more than other. Bulding of an experimental setup to be able to drop all the batteries at the same time and from the same height. Shows differences in texture of the inside components of a charged and empty battery, mentions that the bouncing is a result of the gel-like texture of the anode part of the battery, which solidify when the battery is empty. specifies that this test is only relevant for alkaline batteries.

##### Already existing scientific studies/theories :
1. « Do dead batteries really bounce and charged ones not », Ada McVean B. C., https://www.mcgill.ca/oss/article/did-you-know-general-science/do-dead-batteries-really-bounce-and-charged-ones-not \

16h : Found out that the discharge time of a battery is long (takes more than 3 hours). It will be necessary to come during the week to be able to discharge several batteries for varying lengths of time and have the possibility to run tests \
16h30 : Better understanding of the discharge device, deductions based on multiple tests throughout the day:
1. If the current is too high, the discharge stops quickly even though the battery still holds energy (but not enough for the chosen current) $\rightarrow$ therefore, discharges should be done at lower currents to consume more energy. The most suitable current balancing discharge rate and discharge time was set at 0.5 A.
2. The end voltage of 1.0 V is not low enough to satisfactorily discharge the battery; it can be set to 0.1 V before starting the discharge.
3. The most satisfactory procedure chosen is as follows : set the initial parameters such as capacity 1550 mAh, current 0.5 A, end voltage 0.1 V, and start the discharge. Produce different charge states by subjecting various batteries to the discharger for varying lengths of time, so that the consumed capacities (provided by the discharger) increase in increments of 100 mAh. The batteries are then numbered, and their data are entered into a table (see [State Charge Table](../charge_state_table.ipynb)) that will be filled throughout the week with their consumed capacity and discharge time.

#### **13/10/25**
The tuning fork experiment was standardized to minimize variations, using a controlled ball drop setup. The measured decay constants remained consistent across all tests, indicating that damping is dominated by the tuning fork rather than the battery or the box. The energy transfer experiment was therefore deemed inconclusive, and future tests will focus on direct box–battery impact measurements (more details see [Day Report 13/10/25](Kinetic_voltmeter/13-10-25/day_report.ipynb))

#### **14/10/25**
The day was spent testing the tuning fork setup with batteries attached using paper tape. It was observed that the tape alone significantly dampens the sound, and adding the battery further constrained the branch’s vibrations, making reproducible measurements difficult. Pendulum tests showed only slight differences, likely due to rebound effects, and the setup lacked sufficient precision to detect small changes in the battery’s internal structure. Sound impact experiments with batteries on different surfaces produced noisy signals with high variation, although discharged batteries sometimes showed a slightly higher frequency peak. (more details see [Day Report 14/10/25](Kinetic_voltmeter/14-10-25/day_report.ipynb))

#### **15/10/25**
The day focused on several small experimental tests on batteries and tuning forks. The ruler-battery test showed no significant difference between charged and partially discharged batteries, though the setup was rudimentary. The tuning fork test without the box confirmed that the fork itself dominates the recorded sound, with the decay constant about 40% larger than with the box. Additional tests explored impact and vibration transmission in rigidly fixed batteries, suggesting that piezoelectric sensors could capture vibrations, but reproducible impacts and precise modeling remain challenging. (more details see [Day Report 15/10/25](Kinetic_voltmeter/15-10-25/day_report.ipynb))

#### **16/10/25**
The day focused on testing vibration transmission through a horizontally fixed battery with a pendulum in contact on one side and impacts applied on the other. Initial results showed no visible difference between charged and discharged batteries, likely due to shared internal structure and imprecision in the setup. Improvements were attempted by stabilizing the support, applying smaller forces, and recording slow-motion video, but differences remained unclear. Future work may require more controlled impacts, accelerometers or piezoelectric sensors, and consideration of the battery’s mechanical impedance to better quantify vibration transmission. (more details see [Day Report 16/10/25](Kinetic_voltmeter/16-10-25/day_report.ipynb))

#### **17/10/25**
The day focused on testing and analyzing vibration transmission in batteries using piezoelectric sensors and a spectrum analyzer. Experiments with both charged and discharged batteries showed no clear differences, and amplitude variations were found to be largely influenced by the sensor and clamping conditions rather than the batteries themselves. Modifications to the setup, including foam spacers and gain analysis, did not improve reproducibility, and resonance shifts were primarily linked to clamping force. Reflections suggested that the continuous excitation used may be insufficient, and future approaches could involve impulsive excitation or controlled signal amplitude to reveal damping effects related to the internal structure of the batteries. The previous bounce experiment setup was used to conduct an experiment on all the discharged batteries. The experiment consists of dropping batteries with known states of charge from a controlled height onto a calibrated force sensor, with each battery tested twice.(more details see [Day Report 17/10/25](Kinetic_voltmeter/17-10-25/day_report.ipynb))

#### **20,22/10/25**
Simple tests showed no measurable change in gain or phase when varying the excitation amplitude, but a reproducible phase and offset variation was later observed between different batteries, revealing that the receiving piezo’s DC offset depends on the battery’s mechanical stiffness. A reliable measurement method was established by driving the excitation piezo sinusoidally, allowing stable extraction of the receiver offset independent of clamping force. These findings suggest that the effective stiffness and Young’s modulus of the battery may be inferred from the phase response, motivating further controlled measurements to compare experimentally derived and model-based stiffness values. (more details see [Day Report 20,22/10/25](Kinetic_voltmeter/20,22-10-25/day_report.ipynb))

#### **29/10/25**
Throughout the day, the origin of the offset was investigated and it was concluded that the clamping distance alone likely pushes the battery out of its linear elastic regime, which explains why the offset does not vary with additional tightening. By assuming quasi-static conditions and small nanometric deformations, we explored methods to extract an effective local Young’s modulus and an associated density per cell, but the approach fails due to insufficient phase-resolution in the oscilloscope, even over large frequency ranges. Consequently, unless a reliable measurement of wave-propagation speed becomes accessible, only the offset–stiffness correlation remains exploitable, though it does not provide a fully usable material parameter on its own. (more details see [Day Report 29/10/25](Kinetic_voltmeter/29-10-25/day_report.ipynb))

#### **30/10/25**
During the day, the influence of the clamping mechanism on the piezoelectric measurement was analyzed, showing that constant static compression does not affect the measured signal, which depends only on time-varying deformation. This result justified a vertical configuration of the setup, simplifying handling while preserving measurement consistency. Additional tests with non-charged metallic samples ruled out a purely mechanical stiffness-based explanation for the observed signal offset, pointing instead toward an electrical origin linked to the battery voltage. Finally, two alternative approaches based on compressing the battery to measure the Young modulus and based on measuring mechanical impedance via electrically isolated contact and current measurements were proposed for future investigation. (more details see [Day Report 30/10/25](Kinetic_voltmeter/30-10-25/day_report.ipynb))

#### **31/10/25**
During the day, experiments with electrically isolated batteries confirmed that the observed offset was of electrical origin rather than mechanical, and attempts to measure the mechanical impedance using the current setup were inconclusive. The spectrum analyzer revealed no significant differences in resonance, suggesting that the piezo’s impedance is much higher than that of the battery. A first setup to analyse the sound of an impact was made and showing hopeful results. In the end of the day, a series of rice experiments were conducted, testing variations in water content, distribution, and use of a lid, revealing that few holes formed and that moisture distribution remained an issue (beginning of the second IPT problem). Several ideas for future experiments were proposed, including preheating the water and soaking the rice to improve consistency and structure formation. (more details see [Day Report Kinetic Voltmeter 31/10/25](Kinetic_voltmeter/31-10-25/day_report.ipynb) and [Day Report Rice Holes 31/10/25](Rice_holes/31-10-25/day_report.ipynb))

#### **07/11/25**
During the day, a series of cooking experiments was conducted to reproduce and analyse the formation of vertical holes (“chimneys”) in cooked rice, using as a reference a successful preparation with long-grain parboiled rice cooked under a lid. The influence of key parameters—including rice type, rice–water ratio, total mass (system thickness), heating protocol, and lid usage—was systematically investigated through successive trials. It was found that long-grain parboiled rice robustly produces uniformly distributed holes, whose visibility increases with cooking time and system thickness, whereas risotto rice tends to inhibit hole formation due to the development of a starch-rich surface film. Additional experiments perturbing the boiling phase support the hypothesis that hole formation is driven by sustained bubble escape from the surface. (more details see [Day Report 07/11/25](Rice_holes/07-11-25/day_report.ipynb))

#### **12/11/25**
Data analysis of the rebound experiment was carried out, yielding first quantitative results that show the expected trend of increasing restitution (elasticity) with battery discharge, through the extraction of the restitution coefficient from piezoelectric timing measurements. In parallel, a theoretical framework was developed to relate mechanical vibration damping and energy dissipation to changes in the internal structure of alkaline batteries during discharge. Several experimental concepts were examined, including tuning-fork–based methods and piezoelectric sensing, highlighting practical limitations of resonance-based approaches and motivating impedance- and transmission-based measurements. Finally, an alternative hypothesis based on impact-generated sound was formulated, linking the partition of impact energy to the discharge-dependent dissipation properties of molecular bonds. (more details see [Day Report 12/11/25](Kinetic_voltmeter/12-11-25/day_report.ipynb))

#### **14/11/25**
The day began with theoretical reflections and handwritten notes aimed at clarifying the physical mechanisms underlying hole formation in cooked rice and identifying future experimental tests. Systematic experiments were then conducted to assess the role of boiling intensity, demonstrating a strong correlation between sustained boiling, bubble formation, and the emergence of holes, while reduced or suppressed boiling significantly diminished the effect. In the afternoon, the influence of rice type and preparation was investigated, showing that maintaining boiling enables hole formation even in basmati rice, though with reduced visibility compared to long-grain rice. Overall, the results confirm boiling as a key control parameter, while also indicating secondary dependencies on rice type, geometry, and concentration, with the specific role of starch concentration remaining inconclusive. (more details see [Day Report 14/11/25](Rice_holes/14-11-25/day_report.ipynb))

#### **20/11/25**
At 9 the first IPT fight simulation was done with the groups of "Crazy cylinder" and "Electrostatic Cyclotron". \
The day was then dedicated to revisiting and refining the data analysis of the rebound experiment, with a reassessment of the assumption of a constant restitution coefficient across successive impacts. The analysis revealed that the restitution coefficient increases with the number of rebounds, indicating progressively more elastic collisions and motivating a focus on the first rebound as the most relevant indicator of energy dissipation. To clarify this behavior, the evolution of the mechanical power prior to each rebound was evaluated, revealing the existence of two distinct energy regimes. The results suggest that at lower energies, damping mechanisms associated with zinc particle displacement become negligible, and the system behaves increasingly like an ordinary elastic solid. (more details see [Day Report 20/11/25](Kinetic_voltmeter/20-11-25/day_report.ipynb))

#### **21/11/25**
The day focused on refining the physical interpretation of low-energy wave transmission in batteries, supported by compression attempts and longitudinal cutting, which revealed clear structural differences between charged and discharged anodes. These observations suggest that, at low energies, vibrations are mainly transmitted by the global battery structure rather than dissipated in the anode. Further work was dedicated to restoring and improving the acoustic measurement setup, although previously observed spectral features could not be reproduced, and a tested new spectrum analyzer proved unsuitable. The day ended with an IPT meeting to select the problems to be presented and discussed. Ghita proposed a new approach using machine learning to classify batteries from impact sounds, for which an improved experimental protocol and a diverse set of charged and discharged batteries were planned. (more details see [Day Report 21/11/25](Kinetic_voltmeter/21-11-25/day_report.ipynb))

#### **24/11/25**
During the day, gain curves were measured using different piezoelectric sensors and clamping conditions, and analyzed within the framework of a damped forced oscillator to extract quality factors. A clear trend was observed in the low-energy regime, showing increased damping for more discharged batteries, suggesting that charged and discharged states can already be distinguished at low excitation levels. These results motivate further reproducibility tests and a refined physical interpretation of the resonances, as well as future measurements at higher energy using signal amplification. (more details see [Day Report 24/11/25](Kinetic_voltmeter/24-11-25/day_report.ipynb))

#### **27/11/25**

Recordings were captured with a smartphone and saved via Audacity, encompassing 28 discharged batteries of various brands and states of discharge, and 19 charged batteries. (more details see [Day Report 27/11/25](Kinetic_voltmeter/27-11-25/day_report.ipynb))

#### **Last weeks**
Over the past two weeks, efforts were focused on refining piezoelectric resonance measurements and developing a machine learning approach to classify battery charge states. Piezo experiments confirmed a decrease in quality factor with battery discharge in the low-energy regime, consistent with the two-energy-level hypothesis, and potential mechanisms for increased damping were qualitatively proposed. Acoustic measurements of controlled battery discharges were collected, and a machine learning model using MFCC and Contr coefficients was successfully trained, achieving 92% accuracy in distinguishing charged and discharged batteries. These results provide a foundation for future work, including regression analysis to quantify state of charge and further exploration of high-energy phenomena such as zinc-grain damping. (more details see [Weeks report](Kinetic_voltmeter/Last_weeks/weeks_report.ipynb))
